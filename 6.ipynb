```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
c2=2.92e4
c3=7.31e6 #s

def tcoll(a):
    return c3*a**(13/4) *np.exp(c2*a)
```

```python
a=np.linspace(0,1,100000)
plt.plot(a, np.log(tcoll(a)), label="Collision time $t_{coll}$")
plt.title("Collision time with respect to the scale factor")
plt.xlabel("Scale factor a(t)")
plt.ylabel("Collision time ln($t_{coll}$) [s]")
plt.savefig("Collisiontime.pdf")
plt.xscale("log")
plt.legend(loc="best")
```

```python
def tau(z):
    return 4*c5/3 *(1+z)**(3/4) *np.exp(-c2/(1+z))
c5=8.8e8
def a(z):
    return 1/(1+z)
```

```python
def p(z):
    return a(z)*tau(z)*np.exp(-tau(z))*(3/4+c2*a(z))
```

```python
z=np.linspace(0, 1500,1500)
```

```python
plt.plot(z,p(z))
plt.xlabel("Redshift z")
plt.ylabel("P(z)")
plt.title("Probability distribution for last scattering at redshift z")
plt.savefig("ProbabilityLastScattering.pdf")
```
